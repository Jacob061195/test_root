```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
```

```python
# 1. Charger un signal audio (ou créer un signal de test)
# Exemple avec un signal sinusoïdal variant dans le temps
sr = 22050  # Taux d’échantillonnage
t = np.linspace(0, 2, 2 * sr, endpoint=False)
signal = np.sin(2 * np.pi * 220 * t) + np.sin(2 * np.pi * 440 * t * (t < 1))
```

```python
# 2. Appliquer la STFT
D = librosa.stft(signal, n_fft=1024, hop_length=256, window='hann')

# 3. Convertir en échelle dB pour affichage
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)
```

```python
# 4. Affichage du spectrogramme
plt.figure(figsize=(10, 5))
librosa.display.specshow(S_db, sr=sr, hop_length=256, x_axis='time', y_axis='hz', cmap='viridis')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogramme (STFT)')
plt.tight_layout()
plt.show()
```
