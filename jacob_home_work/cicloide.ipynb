```python
import numpy as np
import matplotlib.pyplot as plt

# Param√®tres
r = 1  # Rayon du cercle
t = np.linspace(0, 2 * np.pi, 1000)  # Param√®tre t de 0 √† 2œÄ

# √âquations param√©triques de la cyclo√Øde
x = r * (t - np.sin(t))  # x(t) = r(t - sin(t))
y = r * (1 - np.cos(t))  # y(t) = r(1 - cos(t))

# Calcul de l'aire sous la courbe en utilisant la m√©thode des trap√®zes
# L'aire est l'int√©grale de y(t) * dx/dt dt
dx_dt = r * (1 - np.cos(t))  # dx/dt = r(1 - cos(t))
integrand = y * dx_dt  # y(t) * dx/dt

# Int√©gration num√©rique avec la m√©thode des trap√®zes
A = np.trapz(integrand, t)  # Aire sous la courbe

# Affichage des r√©sultats
print(f"Aire sous la cyclo√Øde pour un cycle complet : A = {A:.4f}")

# Trac√© de la cyclo√Øde
plt.figure(figsize=(8, 4))
plt.plot(x, y, label="Cyclo√Øde")
plt.fill_between(x, y, color="lightblue", alpha=0.5, label="Aire sous la courbe")
plt.title("Cyclo√Øde et aire sous la courbe")
plt.xlabel("x(t)")
plt.ylabel("y(t)")
plt.legend()
plt.grid(True)
plt.show()
```
