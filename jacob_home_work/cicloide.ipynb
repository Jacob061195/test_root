```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres
r = 1  # Rayon du cercle
t = np.linspace(0, 2 * np.pi, 1000)  # Paramètre t de 0 à 2π

# Équations paramétriques de la cycloïde
x = r * (t - np.sin(t))  # x(t) = r(t - sin(t))
y = r * (1 - np.cos(t))  # y(t) = r(1 - cos(t))

# Calcul de l'aire sous la courbe en utilisant la méthode des trapèzes
# L'aire est l'intégrale de y(t) * dx/dt dt
dx_dt = r * (1 - np.cos(t))  # dx/dt = r(1 - cos(t))
integrand = y * dx_dt  # y(t) * dx/dt

# Intégration numérique avec la méthode des trapèzes
A = np.trapz(integrand, t)  # Aire sous la courbe

# Affichage des résultats
print(f"Aire sous la cycloïde pour un cycle complet : A = {A:.4f}")

# Tracé de la cycloïde
plt.figure(figsize=(8, 4))
plt.plot(x, y, label="Cycloïde")
plt.fill_between(x, y, color="lightblue", alpha=0.5, label="Aire sous la courbe")
plt.title("Cycloïde et aire sous la courbe")
plt.xlabel("x(t)")
plt.ylabel("y(t)")
plt.legend()
plt.grid(True)
plt.show()
```
