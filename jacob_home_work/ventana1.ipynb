```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
```

```python
# 1. Créer un signal avec deux fréquences
sr = 1000  # échantillons/seconde
T = 1.0  # durée en secondes
t = np.linspace(0.0, T, int(sr * T), endpoint=False)
signal = np.sin(2 * np.pi * 50 * t) + np.sin(2 * np.pi * 150 * t)
```

```python
# 2. Paramètres de fenêtre
window_size = 200
hop_size = 100
```

```python
# 3. Initialisation pour visualisation
fig, axs = plt.subplots(3, 1, figsize=(12, 10))
```

```python
# 4. Affichage du signal original
axs[0].plot(t, signal)
axs[0].set_title("Signal original dans le temps")
axs[0].set_xlabel("Temps (s)")
axs[0].set_ylabel("Amplitude")
```

```python
# 5. Diviser en fenêtres et calculer FFT de chaque
axs[1].set_title("Fenêtres dans le signal")
axs[2].set_title("Spectre de fréquence par fenêtre")

for i in range(0, len(signal) - window_size, hop_size):
    window = signal[i:i + window_size]
    time_window = t[i:i + window_size]
    
    # Tracer la fenêtre sur le signal
    axs[1].plot(time_window, window, alpha=0.6)
```
