```python
import numpy as np
import matplotlib.pyplot as plt

# Définir la fonction
def f(theta):
    return np.sin(theta)

# Créer un tableau de valeurs pour theta
theta = np.linspace(0, 2 * np.pi, 1000)  # De 0 à 2π avec 1000 points
y = f(theta)

# Tracer la courbe
plt.plot(theta, y, label=r'$\sin(\theta)$', color='blue')

# Remplir l'aire sous la courbe entre 0 et π
plt.fill_between(theta, y, where=(theta >= 0) & (theta <= np.pi), color='lightblue', alpha=0.5, label='Aire sous la courbe')

# Ajouter des labels et un titre
plt.xlabel(r'$\theta$')
plt.ylabel(r'$\sin(\theta)$')
plt.title('Représentation graphique de $\sin(\\theta)$')
plt.axhline(0, color='black', linewidth=0.5)  # Axe des x
plt.axvline(0, color='black', linewidth=0.5)  # Axe des y
plt.legend()

# Afficher le graphique
plt.grid(True)
plt.show()
```
