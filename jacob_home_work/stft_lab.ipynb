```python
import matplotlib.pyplot as plt
from scipy.signal import spectrogram
from scipy.io import wavfile
import numpy as np
```

```python
fs_downloaded, data_downloaded = wavfile.read(r"c:\Users\rodri\Downloads\Beethoven_Diabelli_Variation_No._13.wav")
```

```python
data_downloaded.shape
```

```python
if len(data_downloaded.shape) > 1:
    data_downloaded = data_downloaded[:,0]
```

```python
freq_d, time_d, Sxx_d = spectrogram(data_downloaded, fs_downloaded)
```

```python
plt.pcolormesh(time_d, freq_d, 10 * np.log10(Sxx_d), shading="gouraud")
plt.ylabel("Frecuencia (Hz)")
plt.xlabel("Tiempo (s)")
plt.title("Espectograma")
plt.colorbar(label="Intensidad (dB)")
plt.show()
```
